```python
import pandas as pd  


df = pd.read_parquet(r"C:\LibriSpeech\features.parquet", engine="pyarrow")  
 
print("Dataset Info:")
print(df.info())

  
print("\nFirst 5 Rows:")
print(df.head())

 
print("\nMissing Values:")
print(df.isnull().sum())

 
print("\nSummary Statistics:")
print(df.describe())


if "Speaker_ID" in df.columns:
    print("\nSpeaker Distribution:")
    print(df["Speaker_ID"].value_counts())


if "Chapter_ID" in df.columns:
    print("\nChapter Distribution:")
    print(df["Chapter_ID"].value_counts())


numeric_df = df.select_dtypes(include=['float64', 'int64'])  
print("\nFeature Correlation:")
print(numeric_df.corr())



df.to_csv("C:/Users/opsb2/ammu/features.csv", index=False)
print("\nCSV file saved.")
```
